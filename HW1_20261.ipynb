Suppose $\theta$ is a random variable generated from a beta distribution as: $\theta \sim \text{Beta}(a^2,b)$. Also assume that  the expectation of $\theta$ is $m$: $E[\theta] = m$

For more information about the $\text{Beta}$ distribution see https://en.wikipedia.org/wiki/Beta_distribution.

First we solve for $b$ in terms of $a^2$ and $m$

$$
m=\frac{a^2}{a^2+b}
$$

$$
m(a^2+b)=a^2
$$

$$
ma^2 + bm=a^2
$$

$$
b = \frac{a^2-ma^2}{m}
$$

$$
b = \frac{a^2(1-m)}{m}
$$

Now we solve for $a$ using the variance formula

$$
v=\frac{a^2b}{(a^2+b)^2(a^2+b+1)}
$$

We can see that $a^2 + b = \frac{a^2}{m}$ because

$$
\begin{aligned}
a^2 + a^2\left(\frac{1 - m}{m}\right)
&= \frac{m a^2}{m} + \frac{a^2 - m a^2}{m} \
&= \frac{m a^2 + a^2 - m a^2}{m} \
&= \frac{a^2}{m}
\end{aligned}
$$

substituting it in $v$ we get the following

$$
v=\frac{a^2 \cdot \frac{a^2(1-m)}{m}}{\left(\frac{a^2}{m}\right)^2\left(\frac{a^2}{m}+1\right)}
$$

Simplifying, we get

$$
v=\frac{a^4 \cdot \frac{(1-m)}{m}}{\left(\frac{a^4}{m^2}\right)\left(\frac{a^2}{m}+1\right)}
$$

$$
v=\frac{\frac{(1-m)}{m}}{\left(\frac{1}{m^2}\right)\left(\frac{a^2}{m}+1\right)}
$$

$$
v=\frac{(1-m)}{\left(\frac{1}{m}\right)\left(\frac{a^2}{m}+1\right)}
$$

$$
\frac{v}{m}=\frac{(1-m)}{\left(\frac{a^2}{m}+1\right)}
$$

$$
v=\frac{m(1-m)}{\left(\frac{a^2}{m}+1\right)}
$$

$$
v=\frac{m-m^2}{\left(\frac{a^2}{m}+1\right)}
$$

$$
v\left(\frac{a^2}{m}+1\right)=m-m^2
$$

$$
\frac{va^2}{m}+v=m-m^2
$$

$$
va^2=m(m-m^2-v)
$$

$$
a^2=\frac{m(m-m^2-v)}{v}

$$
